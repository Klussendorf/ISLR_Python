```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load Advertising dataset
ad = pd.read_csv('Dataset/Advertising.csv').drop('Unnamed: 0', axis=1)
ad.head()
```

```python
ad.info()
```

```python
#figure 2.1  the plot of sales against other predictors in the dataset
fig, (ax1, ax2, ax3) =plt.subplots(1,3,figsize=(10,4))
ax1.scatter('TV','Sales', data= ad, alpha= 0.5)
x = np.arange(300)
m, b = np.polyfit(ad.TV, ad.Sales, 1)
y1= m*x+b
ax1.plot(x,y1, color ='red')
ax1.set_ylabel('Sales')
ax1.set_xlabel('TV')


ax2.scatter('Radio','Sales',data=ad, alpha=0.5)
m2,b2=np.polyfit(ad.Radio, ad.Sales, 1)
x2 = np.arange(50)
y2= m2*x2+b2
ax2.plot(x2,y2, color='red')
ax2.set_ylabel('Sales')
ax2.set_xlabel('Radio')


ax3.scatter('Newspaper', 'Sales',data=ad, alpha= 0.5)
m3,b3=np.polyfit(ad.Newspaper, ad.Sales, 1)
x3 = np.arange(100)
y3= m3*x3+b3
ax3.plot(x3,y3, color='red')
ax3.set_ylabel('Sales')
ax3.set_xlabel('Newspaper')
```

## Applied Exercise using College dataset

```python
college= pd.read_csv('/Users/YaoJunyan/Documents/ISLR_Python/Dataset/College.csv').drop('Unnamed: 0',axis=1)
college.info()
```

```python
college.describe()
```

```python
college['private'] = pd.factorize(college.Private)[0]
college.private.value_counts()
```

```python
#for space concern, i am only displaying the first 7 columns, you can always add more columns in the vars list
sns.pairplot(college, vars= ['private','Apps' ,'Accept', 'Enroll', 'Top10perc','Top25perc', 'F.Undergrad'])
```

```python
sns.boxplot(x='Private', y='Outstate', data= college)
```

```python

college['Elite'] = np.where(college['Top10perc']>50, 'yes', 'no')
college['Elite'].describe()
college['Elite'].value_counts()
```

```python
sns.distplot(college.Accept, kde= False)
sns.distplot(college.Enroll, kde= False)
```
